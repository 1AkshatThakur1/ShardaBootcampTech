##**Univariate Analysis**

###**Q1- Which gender had a greater presence among the passengers?**

```python
sns.countplot(data=df, x='sex')
```

###**Q2 - Which passenger class had the fewest number of passengers?**

```python
sns.countplot(data=df, x='class')
```

###**Q3 - Was the number of passengers who survived greater or less than the number of passengers who did not survive?**

```python
sns.countplot(data=df, x='survived')
```

###**Q4 - From which town most number of passengers embarked?**

```python
sns.countplot(data=df, x='embark_town')
```

###**Q5 - How does the survival rate differ between males and females?**

```python
sns.countplot(data=df, x='sex', hue='survived')
```

###**Q6 - Which class passengers survived more in number?**





```python
sns.countplot(data=df, x='class', hue='survived')
```

###**Q7 - Average fare was high for which gender?**

```python
import numpy as np
```

```python
sns.barplot(data = df, x ='sex', y = 'fare', estimator = np.mean)
```

###**Q8 - Explore the relationship between fare and survival. Do passengers who paid higher fares have higher chances of survival?**

```python
sns.boxplot(data=df, x='survived', y='fare')
```
