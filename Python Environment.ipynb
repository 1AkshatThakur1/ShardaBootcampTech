##**Write a Python expression that calculates the area of rectangle. It should ask to input the length and width.**

```python
breadth = float(input("Enter the breadth(in cm): "))
length = float(input("Enter the length(in cm): "))
area = breadth * length
print("The area of rectangle is: ",area,"cm")
```
